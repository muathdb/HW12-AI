# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (you can substitute with actual values if desired)
data = {
    'Department': ['A', 'B', 'C', 'D', 'E'],
    '2019': [55, 40, 85, 30, 75],
    '2024': [60, 50, 65, 45, 90]
}

df = pd.DataFrame(data)

# Plot setup
fig, ax = plt.subplots(figsize=(10, 6))

# Plot lines and text labels
for i in range(len(df)):
    ax.plot([1, 2], [df['2019'][i], df['2024'][i]], marker='o', label=df['Department'][i])
    ax.text(1 - 0.05, df['2019'][i], f"{df['Department'][i]}: {df['2019'][i]}", ha='right', va='center')
    ax.text(2 + 0.05, df['2024'][i], f"{df['2024'][i]}", ha='left', va='center')

# Formatting
ax.set_xlim(0.8, 2.2)
ax.set_xticks([1, 2])
ax.set_xticklabels(['2019', '2024'])
ax.set_title('Change in Department Revenue (2019 vs 2024)', fontsize=14)
ax.axvline(1, color='gray', linestyle='--', linewidth=0.5)
ax.axvline(2, color='gray', linestyle='--', linewidth=0.5)
ax.tick_params(left=False, bottom=False)
ax.set_yticks([])
ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)

plt.tight_layout()
plt.show()
```
